```python
# find number of movies  starting with A for each group
def func(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(func)
```

```python
# find ranking of each movie in the group according to IMDB Score
def rank_movie(group):
  group['genre_ranking'] = group['IMDB_Rating'].rank(ascending = False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
def normal(group):
    min_rating = group['IMDB_Rating'].min()
    max_rating = group['IMDB_Rating'].max()
    # Use an if-else structure to correctly handle cases where max_rating - min_rating is zero
    if (max_rating - min_rating) == 0:
        group['normalized_rating'] = 0.0
    else:
        group['normalized_rating'] = (group['IMDB_Rating'] - min_rating) / (max_rating - min_rating)
    return group

genres.apply(normal)
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head(2)
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
ipl.head(2)
```

```python
# find the batsman with max no of six
six = ipl[ipl['batsman_runs'] == 6]

six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=0)
```

```python
# find batsman with most number of 4's and 6's last 5 over
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]

# temp_df.value_counts('batsman')
temp_df.groupby('batsman')['batsman_runs'].count().sort_values(ascending=0).head(1).index[0]
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# create a function that can return the highest score of any batsman

temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=0).head(1).values[0]
```

```python
def high_score(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]

  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=0).head(1).values[0]
```

```python
high_score('DA Warner')
```
