### What is Pandas

built on top of the Python programming language.



### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

```python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]

runs_ser = pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name
marks = pd.Series(marks,index=subjects,name='Nitish ke marks')
marks
```

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name='yogesh ke marks')
marks_series
```

### Some Important Series Methods

```python
# astype
# between
# clip
# drop_duplicates
# isnull
# dropna
# fillna
# isin
# apply
# copy
```

```python
# astype
import sys
sys.getsizeof(vk)
```

```python
sys.getsizeof(vk.astype('int16'))
```

```python
# between
vk[vk['runs'].between(51, 99)].size
```

```python
# clip
subs
```

```python
subs.clip(100,200)
```

```python
# drop_duplicates
temp = pd.Series([1,1,2,2,3,3,4,4])
temp
```

```python
temp.drop_duplicates(keep='last')
```

```python
temp.duplicated().sum()
```

```python
# isnull
temp.isnull().sum()
```

```python
# dropna
temp.dropna()
```

```python
# fillna
temp.fillna(temp.mean())
```

```python
# isin
vk[(vk == 49) | (vk == 99)]
```

```python
vk[vk.isin([49,99])]
```

```python
# apply
movies
```

```python
# apply
# Select the specific column first, then apply the string operation
movies['lead'].apply(lambda x: x.split()[0].upper())
```

```python
# copy
# Creating a copy of the series to avoid modifying the original data
vk_copy = vk.copy()
vk_copy.iloc[0] = 1000

print("Original first value:", vk.iloc[0].values[0])
print("Copy first value:", vk_copy.iloc[0].values[0])
```
